Beste bide da: Routh-Hurwitzen irizpidea erabiltzea. Kalkulu sinbolikoa erabiltzeko sympy libreraia erabiko dugu.

```python
import sympy as syp
s=syp.var('s')
K=syp.var('K')
g = K*1000/(s**3+15*s**2+50*s+K*1000)
```

```python
def routhHurwitz(Ts,*args):
    den = syp.Poly(Ts.as_numer_denom()[1],s).all_coeffs()
    n = len(den)-1
    if n < 2:
        return None
    m = syp.matrices.zeros(n+1)
    # Insert the first two rows
    c = [den[i] for i in range(0,n+1,2)]
    for i in range(0,len(c)):
        m[0,i] = c[i]
    c = [den[i] for i in range(1,n+1,2)]
    for i in range(0,len(c)):
        m[1,i] = c[i]
  
    # Compute the rest of the entries
    for i in range(2,n+1):
        for j in range(1,n):
            if m[i-1,j] == 0:
                m[i,j-1] = m[i-2,j]
            elif m[i-1,0] !=0:
                m[i,j-1] = ((m[i-1,0]*m[i-2,j]-m[i-2,0]*m[i-1,j])/m[i-1,0]).together().simplify()
          
      
    m = m.col_insert(0, syp.Matrix(([s**(n-i) for i in range(0,n+1)])))
    return m
```
