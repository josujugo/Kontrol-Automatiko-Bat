Izan bedi berrelikadura unitarioko kontrol sistema bat, kontrolagailua G_{c}(s)=K izanik. Sistema fase minimokoa da. K=1 denean, begizta irekiko transferentzia funtzioaren Bode diagrama hau da: 



(a) Aztertu begizta itxiko egonkortasuna.

(b) Aztertu errore-seinalearen balioa maila sarreraren aurrean K-ren arabera.

(c) Kontrolagailua era digitalean inplementatzen bada, aurreko erantzuna berdina izango litzateke?

#### Ebazpena

$P=0$ dela jakinda, hiru datu garrantzitsu daude grafikoan:

- Irabazia $\omega\to 0$ doanean, $K_{p_0}\simeq 10^{6/20}$
- Irabazi  margina IM $\simeq 15 db$
- Fase margina FM $\simeq 100$

  a) Sistema egonkorra da begizta itxian Irabazi margina ez bada gainditzen. AM eta FM positiboak direlako, $K<10^{IM/20}=10^{15/20}$ denean.

  b) Sistema 0 motakoa da ($K_{p_0}$ finitua da). Orduan, errore seinalea
  $e_{ss}=\frac{1}{1+K_p}$, $K_p= K_{p_0}*K$, $K<10^{IM/20}=10^{15/20}$ denean (sistema egonkorra).

  c) Prezisioa ez da aldatzen diskretizazioagtik, irabazia ez delakoaldatze, sistema egonkorra den bitartean. Hori bai, diskretizazioagatik, fasea distorsionatzen da, eta FM txikiagoa da (eta IM), sistema lehengo ezegonkorra bihurtuz. Lagintze-periodoa txikituz egoera hobea izango da, baina kasu jarraituan baino txarragoa.
