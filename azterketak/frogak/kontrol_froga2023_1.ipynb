2. Aztertu hurrengo sistemen egonkortasuna begizta itxian Nyquisten irizpidea erabiliz, K-ren arabera. Adierazi behar da kasu bietan N, Z eta P-ren balioak eta jatorrian poloak dauden ala ez.

a) Har dezagun ohiko sistema berrelikatu bat, $G_{p}(s)=\frac{20(s+20)}{(s+2)(s^{2}+10s+100)}$, $H(s)=\frac{1}{1+s/100}$ eta $G_{c}=k$, izanik.

(b) Har dezagun berriro $G_{c}=k$, orain $G_{p}(s)=\frac{s-0.5}{s(s+2)^{2}}$ eta $H(s)=1$, izanik


##### Emaitza

Hasten da a) kasuarekin. Gogoratu behar da, egonkortasun azterketa begitz irekiko transferentzia funtzioarekin egiten dela 

$G(s) H(s)=\frac{20(s+20)}{(s+2)(s^{2}+10s+100)(1+s/100)}$


Beraz, $P=0$ eta jatorrizko poloen kopurua 0 ere (0 motako sistema). Orain, Nyquist diagramak eta EKak marraztu daitezke.

```scilab
%plot inline
s=%s;


g=20*(s+20)/(s+2)/(s^2+10*s+100);g.dt='c';
h=1/(1+s/100);
scf(1);
nyquist(g*h)

scf(2);
evans(g*h)
```

Nyquist aztertzean ikustean da K txikieantzat $N=0$ dela, beraz, $Z=0$ eta sistema egonkorra begizta itxian.

Handitzean K, gaindituz $1/0.2=5$, $N=2$, $Z=2$, bi polo ezezgonkor beraz.

EKa aztertuta ondorio bera lortzen da.
____________________________________________

b) aztertzen dugu orain.

Ikusten da $P=0$ ere, baina jatorrian polo bat dago (1 motako sistema).

```scilab
%plot inline
s=%s;

g1=(s-0.5)/s/(s+2)/(s+2);
h1=1;g1.dt='c';

scf(1);
nyquist(g1*h1)

scf(2);
evans(g1*h1)
```

Orain Nyquist aztertzean gogoratu behar da infinituan buelta erdi bat ematen dela, erloju orratzen norazkoan, maiztasun negatiboetan hasiz (behean) eta po.sitiboetan amaituz (goian). Zoom egiten ikusten da ez dela mozten ardatz errela eta beti egoera berbera da $N=+1$, hau da $Z=1$.

EKa aztertzean ondorio bera lortu daiteke.

Ez da normalean egiten ,baina kasu honetan aztertu daiteke ze gertatzen da K negatiboetan. Nyquist diagraman aztertuz, K negatibo handietan inguratzen da -1/K puntua (ardatz erreal positiboan) eta $N=2$, baina K txikituz jada ez da inguratzen (hurruntzen da 1/K puntua eskumatik), $N=0$ eta sistema egonkorra izango da. Ondorio bera ikusi zitekeen EKn

```scilab
%plot inline
evans(-g1*h1,40)
```
